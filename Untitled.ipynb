```python

import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email import encoders
import email.message
import pandas as pd

# Configure o número da porta e o nome do servidor
smtp_port = 587                 # Standard secure SMTP port
smtp_server = "smtp.gmail.com"  # Google SMTP Server

# Configure as listas de e-mail
email_from = "eel2000martins@gmail.com"
email_list = [
     ['dianaabilheira@gmail.com', 'Diana Abilheira', 'prostibulo del Pueblo'],
     ['diana-abilheira@hotmail.com', 'Diana Abilheira', 'prostibulo del Pueblo'],
    
]

# Defina a senha (melhor referenciar externamente)
pswd = "xggfanvncxmdsrqn" # As shown in the video this password is now dead, left in as example only


# nomeie o assunto do e-mail
subject = "Sorry we missed you at Web Summit"


# Defina a função de e-mail (não chame de e-mail!)
def send_emails(email_list):

    for person in email_list:

        # Faça o corpo do e-mail
        body = f"""Hi {person[1]},
Unfortunately, we didn't have time to meet at the Web Summit.

What your team is doing at {person[2]} is awe-inspiring. I want to introduce myself. My name is Joel. I work for a company called Bold Ideas. We help startups and companies to reach the next level. 

You can find more information here:
https://websummit.bold-ideas.pt

We would love to help you with any web-related work, so do not hesitate to contact me if you have any requirements or want guidance or help.

Kind Regards,
Joel Veloso - Bold Ideas
https://bold-ideas.pt
        """

        # faça um objeto MIME para definir partes do e-mail
        msg = MIMEMultipart()
        msg['From'] = email_from
        msg['To'] = person[2]  #com este campo vacio eliminace o BCC 
        msg['Subject'] = subject

        # Anexe o corpo da mensagem
        msg.attach(MIMEText(body, 'plain'))

        # Converte como string
        text = msg.as_string()

        # Conecte-se com o servidor
        print("Connecting to server...")
        TIE_server = smtplib.SMTP(smtp_server, smtp_port)
        TIE_server.starttls()
        TIE_server.login(email_from, pswd)
        print("Succesfully connected to server")
        print()


        # Envie e-mails para "pessoa" conforme a lista é iterada
        print(f"Sending email to: {person}...")
        TIE_server.sendmail(email_from, person, text)
        print(f"Email sent to: {person}")
        print()

    # Fecha a porta
    TIE_server.quit()


# Executa a função
send_emails(email_list)




        




```
